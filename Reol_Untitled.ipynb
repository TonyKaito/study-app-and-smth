```python
### helper function, emulate removing from an array with a static array (the counter var is the end of the array, extra values act as padding (like it is normally in the stack anyways))
### should be done in place, I think Python is passed by reference
def removeFromArray(arr: List[int], arr_counter: int, remove_ind: int) -> None:
  count = 0
  for i in range(0, arr_counter - 1):
    if (i == remove_ind):
      count += 1
    arr[i] = arr[count]
    count += 1

  arr_counter -= 1
```
